```python

# coding: utf-8

# In[93]:

print("LSTM")
import keras
from keras.layers import LSTM,Embedding,Dropout,Activation,Dense,Bidirectional,GRU
from keras.models import Sequential,Input,Model
import numpy as np
from keras.utils import to_categorical
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import  pad_sequences
import time


# In[2]:


start_time = time.time()





# In[3]:


import numpy as np
glove_filename = 'glove.6B.50d.txt'



embeddings = {}

print('Extract %s' % glove_filename)
with open(glove_filename, 'r') as f:
    for line in f:
        values = line.split(' ')
        word = values[0]
        embedding = np.asarray(values[1:],dtype=np.float32)
        embeddings[word] = embedding
        
print('Embeddings size: %d' % len(embeddings))
#print(embeddings)


# In[37]:


import json
def loadDataset(filename, size=100):
    label_category = {
        'neutral': 0,
        'entailment': 1,
        'contradiction': 2
    }
    dataset = []
    sentence1 = []
    sentence2 = []
    labels = []
    with open(filename, 'r') as f:
        i = 0
        not_found = 0
        for line in f:
            row = json.loads(line)
            label = row['gold_label'].strip()
            if label in label_category:
                labels.append(label_category[label])
                sentence1.append(row['sentence1'].strip())
                sentence2.append(row['sentence2'].strip())
    return to_categorical(labels),sentence1,sentence2
                   
            
            
            

labels,train_sentence1,train_sentence2 =  loadDataset('snli_1.0/snli_1.0_dev.jsonl', 100) 


# In[30]:


labels


# In[38]:


NUM_WORDS = len(embeddings) #200000

print('Found %s samples.' % len(train_sentence1))

train_sentences = train_sentence1 + train_sentence2

tokenizer = Tokenizer(num_words = NUM_WORDS)
tokenizer.fit_on_texts(train_sentences)
sentence1_word_sequences = tokenizer.texts_to_sequences(train_sentence1)
sentence2_word_sequences = tokenizer.texts_to_sequences(train_sentence2)
word_index = tokenizer.word_index

print('Found %s unique tokens.' % len(word_index))


# In[40]:


from collections import defaultdict
embedding_dim = 50
words_len = min(NUM_WORDS, len(word_index))
# word_embedding_matrix = np.zeros((words_len + 1, embedding_dim))
word_embedding_matrix = np.random.random((words_len + 1, embedding_dim))
k = 0
for word, i in word_index.items():
    if i >= NUM_WORDS:
        continue
    embedding_vector = embeddings.get(word)
    #print(embedding_vector)
    if embedding_vector is not None:
        word_embedding_matrix[i] = embedding_vector
        k += 1
        
max_word_count_text = 0
text_count = defaultdict(int)
for sentence in sentence1_word_sequences:
    #print('sent',sentence)
    #print(len(sentence))
    max_word_count_text = max(max_word_count_text, len(sentence))
    text_count[len(sentence)] += 1
print(text_count)
max_word_count_hypo = 0
hypo_count = defaultdict(int)
for sentence in sentence2_word_sequences:
    max_word_count_hypo = max(max_word_count_hypo, len(sentence))
    hypo_count[len(sentence)] += 1
  


# In[45]:


max_seq_length = 20
s1_data = pad_sequences(sentence1_word_sequences, maxlen = max_seq_length)
s2_data = pad_sequences(sentence2_word_sequences, maxlen = max_seq_length)



print('Shape of sentence1 tensor:', s1_data.shape)
print('Shape of sentence2 tensor:', s2_data.shape)
print('Shape of label tensor:', labels.shape)
print(s1_data)


# In[94]:


def word_context(input, name):
    embedding = Embedding(words_len + 1,
                     embedding_dim,
                     weights = [word_embedding_matrix],
                     input_length = max_seq_length,
                     trainable = False,
                     name = name + '_embedding')(input)
    
    word = Dropout(0.1)(embedding)

    context =Bidirectional( LSTM(100))(word)
    print("word",word)
    print("condtext",context)
    return (word, context)


# In[96]:


def build_model(input_array):
    sentence1_input = Input(shape=(max_seq_length,), dtype='int32', name = 'text')
    sentence2_input = Input(shape=(max_seq_length,), dtype='int32', name = 'hypothesis')
    
    (text_embedding, text_context) = word_context(sentence1_input, 'text')
    (hypo_embedding, hypo_context) = word_context(sentence2_input, 'hypothesis')
    #e = keras.layers.Concatenate(axis=-1)([text_context,hypo_context])
    s = keras.layers.Concatenate(axis=-1)([text_embedding,hypo_embedding])
    f = Bidirectional(LSTM(200))(s)
    f = Dense(3)(f)
    f = Activation("softmax")(f)
    
    
    #k = LSTM(100)(e)
    
    model = Model(inputs=[sentence1_input,sentence2_input],outputs=f)
    model.compile('rmsprop', 'mse',metrics = ["accuracy"])
    output_array = model.predict(input_array)
    print(output_array.shape)
    return model
#input_array = np.random.randint(1000, size=(32, 10))    
model = build_model([s1_data,s2_data])   
    
    


# In[97]:


model.summary()
model.fit([s1_data, s2_data],
                        labels,
                        epochs = 5,
                        batch_size = 1024)
end_time = time.time()
print('Time: %d' % (end_time - start_time))


# In[ ]:


labels,train_sentence1,train_sentence2 =  loadDataset('snli_1.0_test.jsonl', 100) 
sentence1_word_sequences = tokenizer.texts_to_sequences(train_sentence1)
sentence2_word_sequences = tokenizer.texts_to_sequences(train_sentence2)
s1_data = pad_sequences(sentence1_word_sequences, maxlen = max_seq_length)
s2_data = pad_sequences(sentence2_word_sequences, maxlen = max_seq_length)
loss, accuracy = model.evaluate([s1_data,s2_data],labels)
 
print('loss     = ', loss)
print('accuracy = ', accuracy)
labels,train_sentence1,train_sentence2 =  loadDataset('snli_1.0_test.jsonl', 100) 
sentence1_word_sequences = tokenizer.texts_to_sequences(train_sentence1)
sentence2_word_sequences = tokenizer.texts_to_sequences(train_sentence2)
s1_data = pad_sequences(sentence1_word_sequences, maxlen = max_seq_length)
s2_data = pad_sequences(sentence2_word_sequences, maxlen = max_seq_length)
loss, accuracy = model.evaluate([s1_data,s2_data],labels)
 
print('loss     = ', loss)
print('accuracy = ', accuracy)
y = model.predict([s1_data, s2_data],verbose = 0)
n_values = 3
t = []
c = np.eye(n_values, dtype=int)[np.argmax(y, axis=1)]

import csv

with open("lstmtextsd.csv","w") as df:
    fields = ['sentence1', 'sentence2',"labels","prediction"]
    writer = csv.DictWriter(df, fieldnames = fields)
    writer.writeheader()

    csv = csv.writer(df)
    for i in range(len(train_sentence1)):
        if i < 100000000000000000:
            csv.writerow((train_sentence1[i] ,train_sentence2[i] ,labels[i] ,c[i] , y[i]))
            
         
    df.close()
import pandas as pd
def load_data(a):
    df = pd.read_csv(a)
    
    texts = df['sentence1']
    hypotheses = df['sentence2 ']
    #label = df[2]
    text = []
    hypothesis = []

    
    for i in texts:
        text.append(i)
    for i in hypotheses:
        hypothesis.append(i)
    return text, hypothesis

train_sentence1,train_sentence2 = load_data('activepassive.csv')
sentence1_word_sequences = tokenizer.texts_to_sequences(train_sentence1)
sentence2_word_sequences = tokenizer.texts_to_sequences(train_sentence2)
s1_data = pad_sequences(sentence1_word_sequences, maxlen = max_seq_length)
s2_data = pad_sequences(sentence2_word_sequences, maxlen = max_seq_length)
y = model.predict([s1_data, s2_data],verbose = 0)
n_values = 3
t = []
c = np.eye(n_values, dtype=int)[np.argmax(y, axis=1)]

import csv

with open("randomlstm.csv","w") as df:
    fields = ['sentence1', 'sentence2',"labels","prediction"]
    writer = csv.DictWriter(df, fieldnames = fields)
    writer.writeheader()

    csv = csv.writer(df)
    for i in range(len(train_sentence1)):
        if i < 100000000000000000:
            csv.writerow((train_sentence1[i] ,train_sentence2[i],c[i] , y[i]))
            
         
    df.close()
model_json = model.to_json()
with open("model.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("model.h5")
print("Saved model to disk")
```
